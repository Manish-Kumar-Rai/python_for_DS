```python
tips = sns.load_dataset("tips")
```

```python
tips.head(5)
```

```python
tips.groupby("day").tip.mean()
```

```python
#Univariate -> one variable visualization
```

```python
#strip plot
sns.stripplot(y="tip",data=tips,jitter=True)
plt.ylabel("tips($)")
plt.show()
```

```python
#Grouping with stripplot
sns.stripplot(x="day",y="tip",data=tips,size=3,jitter=True)
plt.ylabel("tips($)")
plt.show()
```

```python
#Swarm plot
```

```python
sns.swarmplot(x="day",y="tip",data=tips,hue="sex")
plt.ylabel("tips($)")
plt.show()
```

```python
#Changing Orientation
sns.swarmplot(y="day",x="tip",data=tips,hue="sex",orient="h")
plt.xlabel("tips($)")
plt.show()
```

```python
#Boxplot
sns.boxplot(x="day",y="tip",data=tips)
plt.ylabel("tips($)")
plt.show()
```

```python
#Violinplot
sns.violinplot(x="day",y="tip",data=tips)
plt.ylabel("tips($)")
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="day",y="tip",data=tips)
plt.ylabel("tips($)")
plt.subplot(1,2,2)
sns.violinplot(x="day",y="tip",data=tips)
plt.ylabel("tips($)")
plt.show()
```

```python
###Combining plot
sns.violinplot(x="day",y="tip",data=tips,inner=None,alpha=0.5,color="lightgray")
sns.stripplot(x="day",y="tip",data=tips,jitter=True,color="black")
plt.ylabel("tips($)")
plt.show()
```

```python
#Bi-variate: "Two" variables data visualization
```

```python
#join plot
sns.jointplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
#Density plot
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
plt.show()
```

```python
#pair plot
sns.pairplot(data=tips,hue="sex")
plt.show()
```

```python
sns.barplot(x="Pclass",y="Survived",hue="Sex",data=train)
```

```python
survived = train.loc[train.Survived == 1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived")
```

```python
survived = train.loc[train.Survived == 0,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived")
```

```python
sns.pairplot(train.dropna())
```

```python
#Pclass vs Survived
grid = sns.FacetGrid(train,col="Survived",row="Pclass",aspect=1.5)
grid.map(plt.hist,"Age",alpha=0.5,bins=20)
grid.add_legend()
```

```python
sns.distplot(a=train[train.Embarked == "C"]["Survived"],kde=False,bins=3)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()

sns.distplot(a=train[train.Embarked == "Q"]["Survived"],kde=False,bins=3)
plt.title("QueensTown")
plt.xticks([0,1])
plt.show()

sns.distplot(a=train[train.Embarked == "S"]["Survived"],kde=False,bins=3)
plt.title("Southampton")
plt.xticks([0,1])
plt.show()
```

```python
figbi, axesbi = plt.subplots(2,4,figsize=(16,10))
train.groupby("Pclass")["Survived"].mean().plot(kind="barh",ax=axesbi[0,0],xlim=[0,1])
train.groupby("SibSp")["Survived"].mean().plot(kind="barh",ax=axesbi[0,1],xlim=[0,1])
train.groupby("Parch")["Survived"].mean().plot(kind="barh",ax=axesbi[0,2],xlim=[0,1])
train.groupby("Sex")["Survived"].mean().plot(kind="barh",ax=axesbi[0,3],xlim=[0,1])
train.groupby("Embarked")["Survived"].mean().plot(kind="barh",ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived",y="Age",ax=axesbi[1,1],data=train)
sns.boxplot(x="Survived",y="Fare",ax=axesbi[1,2],data=train)                                                 
```
